```python
# modulos necesarios
import folium
import pandas
import numpy
import algorithms #modulo local de algoritmos de ordenamiento
import time
```

```python
tabla = pandas.DataFrame(data=None,
                         columns=['Algoritmo',
                                  'Soporta negativos',
                                  'soporta flotantes',
                                  'complejidad O',
                                  'mejor caso',
                                  'peor caso',
                                  'T. n = 10',
                                  'T. n = 100',
                                  'T. n = 1K',
                                  'T. n = 10K'])
```

```python
array_entrada = []
array_entrada.append(numpy.random.randint(100, size=10))
array_entrada.append(numpy.random.randint(100, size=100))
array_entrada.append(numpy.random.randint(100, size=1000))
array_entrada.append(numpy.random.randint(100, size=10000))
```

```python
#heap_sort
heapsort = []
heapsort.append("heapsort")
heapsort.append(True)
heapsort.append(True)
heapsort.append("O(n.log(n))")
heapsort.append("O(n.log(n))")
heapsort.append("O(n.log(n))")
for t in range(4):
    ## hallando la media de tiempo de ejecucion 
    media = 0
    for i in range(5):
        start_time = time.time()
        algorithms.heapSort(array_entrada[t])
        media += time.time() - start_time
    media = media / 5 
    heapsort.append(media)
tabla.loc[len(tabla)] = heapsort
```

```python
#quick_sort
quicksort = []
quicksort.append("quicksort")
quicksort.append(True)
quicksort.append(True)
quicksort.append("O(n.log(n))")
quicksort.append("O(n.log(n))")
quicksort.append("O(n^2)")
for t in range(4):
    ## hallando la media de tiempo de ejecucion 
    media = 0
    for i in range(5):
        start_time = time.time()
        algorithms.quickSort(array_entrada[t])
        media += time.time() - start_time
    media = media / 5 
    quicksort.append(media)
tabla.loc[len(tabla)] = quicksort
```

```python
#counting sort
countingsort = []
countingsort.append("countingsort")
countingsort.append(False)
countingsort.append(False)
countingsort.append("O(n)")
countingsort.append("O(n)")
countingsort.append("O(n)")
for t in range(4):
    ## hallando la media de tiempo de ejecucion 
    try:
        media = 0
        for i in range(5):
            start_time = time.time()
            algorithms.counting_sort(array_entrada[t])
            media += time.time() - start_time
        media = media / 5 
    except:
        media = None
        pass
    countingsort.append(media)
tabla.loc[len(tabla)] = countingsort
```

```python
# radixsort
radixsort = []
radixsort.append("radixsort")
radixsort.append(False)
radixsort.append(True)
radixsort.append("O(n)")
radixsort.append("O(n)")
radixsort.append("O(n)")
for t in range(4):
    ## hallando la media de tiempo de ejecucion 
    media = 0
    for i in range(5):
        start_time = time.time()
        algorithms.radix(array_entrada[t])
        media += time.time() - start_time
    media = media / 5 
    radixsort.append(media)
tabla.loc[len(tabla)] = radixsort
```

```python
#bucket sort
bucketsort = []
bucketsort.append("bucket")
bucketsort.append(True)
bucketsort.append(True)
bucketsort.append("O(n)")
bucketsort.append("O(n)")
bucketsort.append("O(n^2)")
for t in range(4):
    ## hallando la media de tiempo de ejecucion 
    media = 0
    for i in range(5):
        start_time = time.time()
        algorithms.bucket_sort(array_entrada[t])
        media += time.time() - start_time
    media = media / 5 
    bucketsort.append(media)
tabla.loc[len(tabla)] = bucketsort
```

```python
puntos = pandas.read_csv('./geodata.csv').head(20)
puntos
```

```python
puntos['HeapSort(x)'] = algorithms.heapSort(puntos['x'])
puntos['QuickSort(x)'] = algorithms.quickSort(puntos['x'])
puntos['CountingSort(x)'] = algorithms.counting_sort(puntos['x'])
puntos['RadixSort(x)'] = algorithms.radix(puntos['x'])
puntos['BucketSort(x)'] = algorithms.bucket_sort(puntos['x'])
puntos
```

```python
# mapa de tiendas donde el circulo representa un mayor o un menor alcanze de clientes en promedio. 
mapa = folium.Map(location=(puntos.iloc[8][0],puntos.iloc[8][1]),zoom_start=(17))
for p in range(len(puntos)):
    folium.CircleMarker(location=[puntos.iloc[p][0],puntos.iloc[p][1]],
                        radius=puntos.iloc[p][2],
                        color='blue',
                        fill=True,
                        fill_color='skyblue'
                       ).add_to(mapa)
    folium.Marker(
        location=[puntos.iloc[p][0],puntos.iloc[p][1]],
        popup="<b>tiendita N."+str(puntos.iloc[p][2])+"</b>",
        icon=folium.Icon(color='green')
    ).add_to(mapa)
mapa
```
